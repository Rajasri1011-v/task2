```python
import pandas as pd

try:
    df = pd.read_csv('Titanic-Dataset (2).csv')
    display(df.head())
    print(df.shape)
except FileNotFoundError:
    print("Error: 'Titanic-Dataset (2).csv' not found.")
    df = None
except Exception as e:
    print(f"An error occurred: {e}")
    df = None
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Check data types and missing values
print(df.info())
print(df.isnull().sum())

# Descriptive statistics for numerical features
print(df[['Age', 'Fare']].describe())

# Distribution of categorical features
print(df['Survived'].value_counts())
print(df['Pclass'].value_counts())
print(df['Sex'].value_counts())
print(df['Embarked'].value_counts())

# Visualize 'Survived' and 'Pclass' distributions
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
df['Survived'].value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Distribution of Survival')
plt.xlabel('Survived')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
df['Pclass'].value_counts().plot(kind='bar', color=['lightgreen', 'lightcoral', 'lightskyblue'])
plt.title('Distribution of Passenger Class')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# Calculate and visualize correlation between numerical features
plt.figure(figsize=(8, 6))
sns.heatmap(df[['Age', 'Fare', 'SibSp', 'Parch']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
# Impute missing 'Age' values
df['Age'] = df.groupby('Pclass')['Age'].transform(lambda x: x.fillna(x.median()))

# Handle missing 'Embarked' values
most_frequent_embarked = df['Embarked'].mode()[0]
df['Embarked'] = df['Embarked'].fillna(most_frequent_embarked)

# Drop the 'Cabin' column
df = df.drop('Cabin', axis=1)

# Remove duplicates
df.drop_duplicates(inplace=True)

# Cap outliers in 'Fare'
fare_99th = df['Fare'].quantile(0.99)
df['Fare'] = df['Fare'].clip(upper=fare_99th)

# Verify data cleaning
print(df.info())
```

```python
# Calculate survival rates for different passenger characteristics
survival_rate_by_sex = df.groupby('Sex')['Survived'].mean()
survival_rate_by_class = df.groupby('Pclass')['Survived'].mean()
survival_rate_by_embarked = df.groupby('Embarked')['Survived'].mean()

print("Survival Rate by Sex:\n", survival_rate_by_sex)
print("\nSurvival Rate by Passenger Class:\n", survival_rate_by_class)
print("\nSurvival Rate by Embarked:\n", survival_rate_by_embarked)


# Calculate descriptive statistics for numerical variables
age_stats = df.groupby('Survived')['Age'].agg(['mean', 'median', 'std'])
fare_stats = df.groupby('Survived')['Fare'].agg(['mean', 'median', 'std'])


print("\nAge Statistics by Survival:\n", age_stats)
print("\nFare Statistics by Survival:\n", fare_stats)

# Calculate descriptive statistics for subgroups
age_stats_by_class = df.groupby(['Pclass', 'Survived'])['Age'].agg(['mean', 'median', 'std'])
fare_stats_by_class = df.groupby(['Pclass', 'Survived'])['Fare'].agg(['mean', 'median', 'std'])

print("\nAge Statistics by Passenger Class and Survival:\n", age_stats_by_class)
print("\nFare Statistics by Passenger Class and Survival:\n", fare_stats_by_class)
```

```python
import scipy.stats as ss

# Correlation between numerical variables
correlation_matrix = df[['Age', 'Fare', 'SibSp', 'Parch', 'Survived']].corr()
print("Correlation Matrix:\n", correlation_matrix)

# Chi-squared tests between categorical variables
# 'Sex' and 'Survived'
contingency_table = pd.crosstab(df['Sex'], df['Survived'])
chi2, p, dof, expected = ss.chi2_contingency(contingency_table)
print("\nChi-squared test for Sex and Survived:")
print("Chi2:", chi2, "p-value:", p)

# 'Pclass' and 'Survived'
contingency_table = pd.crosstab(df['Pclass'], df['Survived'])
chi2, p, dof, expected = ss.chi2_contingency(contingency_table)
print("\nChi-squared test for Pclass and Survived:")
print("Chi2:", chi2, "p-value:", p)

# 'Embarked' and 'Survived'
contingency_table = pd.crosstab(df['Embarked'], df['Survived'])
chi2, p, dof, expected = ss.chi2_contingency(contingency_table)
print("\nChi-squared test for Embarked and Survived:")
print("Chi2:", chi2, "p-value:", p)

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Survival Rate by Passenger Class
plt.figure(figsize=(8, 6))
sns.barplot(x='Pclass', y='Survived', data=df, palette='viridis')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()

# 2. Age Distribution for Survivors and Non-Survivors
plt.figure(figsize=(8, 6))
sns.boxplot(x='Survived', y='Age', data=df, palette='Set2')
plt.title('Age Distribution by Survival Status')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()

# 3. Fare Distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['Fare'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Fares')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()

# 4. Embarkation Point Distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='Embarked', hue='Survived', data=df, palette='Set1')
plt.title('Embarkation Point Distribution by Survival Status')
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.show()

# 5. Heatmap of Correlation Matrix
plt.figure(figsize=(10, 8))
corr_matrix = df[['Age', 'Fare', 'SibSp', 'Parch', 'Survived']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Age, Fare, SibSp, Parch, and Survived')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Select relevant features
features = df[['Age', 'Pclass', 'Fare']]

# Scale the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Apply K-means clustering
kmeans = KMeans(n_clusters=3, random_state=0)  # Experiment with different n_clusters
df['cluster'] = kmeans.fit_predict(scaled_features)

# Analyze the clusters
print(df.groupby('cluster').agg({'Age': 'mean', 'Pclass': 'mean', 'Fare': 'mean'}))

# Visualize the clusters (example using Age and Fare)
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Fare'], c=df['cluster'], cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('KMeans Clustering of Passengers')
plt.colorbar(label='Cluster')
plt.show()
```
